```python
# FILE: Untitled5_v2.ipynb

import numpy as np
import pandas as pd
from scipy.spatial.distance import euclidean
from collections import defaultdict
from sklearn.model_selection import GroupKFold, StratifiedKFold
from sklearn.metrics import log_loss, accuracy_score
import random
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
from itertools import product
from scipy.stats import zscore
import plotly.express as px
from scipy.optimize import minimize
import seaborn as sns

# ------------------ CCure Normalization Function ------------------

def ccure_normalize(df, feature_cols):
    """
    Perform CCure Normalization on specified feature columns by removing talker-specific means.
    
    Parameters:
    - df: pandas DataFrame to normalize.
    - feature_cols: List of column names to apply normalization.
    
    Returns:
    - Normalized DataFrame.
    """
    for col in feature_cols:
        df[col] = df.groupby('Talker')[col].transform(lambda x: x - x.mean())
    return df

# ------------------ Helper Functions ------------------

def add_noise(features, noise_level):
    """Add Gaussian noise to features based on noise_level."""
    return [f + np.random.normal(0, noise_level * abs(f)) for f in features]

def compute_distance(input_features, exemplar_features, weights, exponent):
    """Compute weighted distance between input and exemplar features."""
    weighted_diff = weights * (np.array(input_features) - np.array(exemplar_features))
    distance = np.sum(np.abs(weighted_diff) ** exponent) ** (1/exponent)
    return distance

def compute_similarity(distance, scaling):
    """Compute similarity based on the modified similarity function."""
    return np.exp(-distance ** scaling)

def predict_posterior(input_features, exemplars, weights, exponent, scaling, prior):
    """Predict posterior probabilities for each category."""
    category_similarities = {}
    total_similarity = 0
    for category, ex_list in exemplars.items():
        sim_sum = sum(
            compute_similarity(
                compute_distance(input_features, ex, weights, exponent),
                scaling
            )
            for ex in ex_list
        )
        category_similarities[category] = sim_sum
        total_similarity += sim_sum
    if total_similarity == 0:
        return None
    posteriors = {cat: (sim_sum * prior.get(cat, 1.0)) for cat, sim_sum in category_similarities.items()}
    total_posterior = sum(posteriors.values())
    if total_posterior == 0:
        return None
    posteriors = {cat: (sim / total_posterior) for cat, sim in posteriors.items()}
    return posteriors

def calculate_log_likelihood(true_label, posterior, classes):
    """Calculate log likelihood for a single prediction."""
    if posterior is None:
        return 0
    prob = posterior.get(true_label, 1e-10)
    prob = max(prob, 1e-10)
    return np.log(prob)

def evaluate_model(train_data, tune_data, scaling, group_kfold, weights, distance_exponent, prior):
    """
    Evaluate the model by computing the mean tuning log likelihood across cross-validation folds.
    
    Parameters:
    - train_data: Training dataset (phonetic data)
    - tune_data: Tuning dataset (perceptual optimization fold)
    - scaling: Scaling factor for similarity computation
    - group_kfold: GroupKFold splitter for cross-validation
    - weights: Array of feature weights [weight_f1, weight_f2]
    - distance_exponent: Exponent for distance computation
    - prior: Prior probabilities for each category
    
    Returns:
    - mean_tune_ll: Mean tuning log likelihood across all cross-validation folds
    """
    print(f"Evaluating model with weights=F1: {weights[0]:.8f}, F2: {weights[1]:.8f}; similarity_scaling: {scaling:.8f}")
    tune_log_likelihoods = []
    fold = 1
    for train_idx, _ in group_kfold.split(train_data, groups=train_data['Talker']):
        print(f"  Processing talker fold {fold}...")
        cv_train = train_data.iloc[train_idx].reset_index(drop=True)
        
        # Train exemplars on cv_train
        exemplars_train = defaultdict(list)
        for _, row in cv_train.iterrows():
            category = row['Vowel']
            features = row[['F1_z', 'F2_z']].tolist()
            noisy = add_noise(features, noise_level)
            if not np.any(np.isnan(noisy)) and not np.any(np.isinf(noisy)):
                exemplars_train[category].append(noisy)
        
        # Evaluate on tuning data
        tune_labels = []
        tune_posteriors = []
        
        for _, row in tune_data.iterrows():
            true = row['Response.Vowel']
            input_feat = row[['F1_z', 'F2_z']].tolist()
            noisy_input = add_noise(input_feat, noise_level)
            if not np.any(np.isnan(noisy_input)) and not np.any(np.isinf(noisy_input)):
                posteriors = predict_posterior(
                    noisy_input, exemplars_train, weights, distance_exponent, scaling, prior
                )
                if posteriors:
                    tune_posteriors.append(posteriors)
                    tune_labels.append(true)
        
        if tune_labels:
            fold_tune_log_likelihood = sum(
                [calculate_log_likelihood(t, p, unique_categories) for t, p in zip(tune_labels, tune_posteriors)]
            )
            tune_log_likelihoods.append(fold_tune_log_likelihood)
            print(f"  Fold {fold} - Tuning Log Likelihood: {fold_tune_log_likelihood:.4f}")
        else:
            print(f"  Fold {fold} - No valid tuning predictions.")
        fold += 1
    
    if tune_log_likelihoods:
        mean_tune_ll = np.mean(tune_log_likelihoods)
        print(f"  Mean Tuning Log Likelihood: {mean_tune_ll:.4f}")
    else:
        mean_tune_ll = np.nan
        print("  No valid tuning log likelihoods across folds.")
    
    return mean_tune_ll

# ------------------ Data Loading and Preprocessing ------------------

print("Loading datasets...")
# Set low_memory=False to suppress DtypeWarning
phonetic_data = pd.read_csv(
    '/scratch/ashar58/RA_Florian/Code/Li_Xie_2020_L1_vowels_statistics_general.csv',
    low_memory=False
)
perceptual_data = pd.read_csv(
    "/scratch/ashar58/RA_Florian/Code/Experiment-NORM-AB-after-exclusions.csv",
    low_memory=False
)
print("Datasets loaded successfully.")

# Modify 'Response.Vowel' in perceptual_data
vowel_mapping = {
    '[u]': 'uw1',
    '[ɪ]': 'ih1',
    '[ʊ]': 'uh1',
    '[i]': 'iy1',
    '[æ]': 'ae1',
    '[ɑ]': 'aa1',
    '[ʌ]': 'ah1',  
    '[ɛ]': 'eh1',
}
perceptual_data['Response.Vowel'] = perceptual_data['Response.Vowel'].map(vowel_mapping)
print("Perceptual vowels mapped successfully.")

# Remove rows with NA in 'Response.Vowel'
initial_perc_rows = perceptual_data.shape[0]
perceptual_data = perceptual_data.dropna(subset=['Response.Vowel'])
final_perc_rows = perceptual_data.shape[0]
rows_dropped = initial_perc_rows - final_perc_rows
print(f"Removed {rows_dropped} rows with NA in 'Response.Vowel'.")

# Define feature columns
phonetic_feature_cols = ['F1_gm', 'F2_gm']  # Phonetic features in Hz
perceptual_feature_cols = ['F1_Hz_r', 'F2_Hz_r']  # Perceptual features in Hz

# Compute z-scores for visualization
print("Computing z-scores for visualization...")
# Combine F1 from phonetic and perceptual data
all_F1 = pd.concat([phonetic_data['F1_gm'], perceptual_data['F1_Hz_r']], axis=0)
all_F2 = pd.concat([phonetic_data['F2_gm'], perceptual_data['F2_Hz_r']], axis=0)
combined = pd.DataFrame({
    'F1_z': zscore(all_F1),
    'F2_z': zscore(all_F2)
})

# Assign z-scores using positional indexing
phonetic_data[['F1_z', 'F2_z']] = combined.iloc[:len(phonetic_data)][['F1_z', 'F2_z']].values
perceptual_data[['F1_z', 'F2_z']] = combined.iloc[len(phonetic_data):][['F1_z', 'F2_z']].values
print("Z-scores computed for visualization.")

# ------------------ Model Definitions ------------------

models = {
    'MODEL-1.1 - Natural': {
        'normalize': False,
        'speech_type': 'natural'
    },
    'MODEL-1.2 - Synthetic': {
        'normalize': False,
        'speech_type': 'synthetic'
    },
    'MODEL-2.1 - Natural': {
        'normalize': True,
        'speech_type': 'natural'
    },
    'MODEL-2.2 - Synthetic': {
        'normalize': True,
        'speech_type': 'synthetic'
    }
}

# Initialize dictionaries to store model results
model_results = {}

# ------------------ Iterate Over Models ------------------

for model_name, model_params in models.items():
    print(f"\n----- Running {model_name} -----")
    
    # Copy datasets to avoid modifying originals
    phonetic_df = phonetic_data.copy()
    perceptual_df = perceptual_data.copy()
    
    # Apply CCure Normalization if required
    if model_params['normalize']:
        print("Applying CCure Normalization...")
        phonetic_df = ccure_normalize(phonetic_df, phonetic_feature_cols)
        perceptual_df = ccure_normalize(perceptual_df, perceptual_feature_cols)
        print("CCure Normalization applied.")
    else:
        print("Skipping CCure Normalization.")
    
    # Select the appropriate speech type data
    print("Splitting perceptual data into natural and synthetic based on 'Experiment' column...")
    if 'Experiment' not in perceptual_df.columns:
        raise KeyError("Column 'Experiment' not found in perceptual_data.")
    natural_data = perceptual_df[perceptual_df['Experiment'] == 'Experiment 1a (natural)'].reset_index(drop=True)
    synthetic_data = perceptual_df[perceptual_df['Experiment'] == 'Experiment 1b (synthesized)'].reset_index(drop=True)
    print(f"Natural data count: {natural_data.shape[0]}")
    print(f"Synthetic data count: {synthetic_data.shape[0]}")
    
    if model_params['speech_type'] == 'natural':
        speech_type_data = natural_data
    elif model_params['speech_type'] == 'synthetic':
        speech_type_data = synthetic_data
    else:
        raise ValueError("Invalid speech type specified.")
    
    # Split speech_type_data into optimization and held-out sets based on 'ParticipantID'
    print("Splitting speech type data into optimization and held-out sets based on 'ParticipantID'...")
    if 'ParticipantID' not in speech_type_data.columns:
        raise KeyError("Column 'ParticipantID' not found in speech_type_data.")
    unique_participants = speech_type_data['ParticipantID'].unique().tolist()
    random.shuffle(unique_participants)
    half = len(unique_participants) // 2
    opt_participants = unique_participants[:half]
    holdout_participants = unique_participants[half:]
    listener_opt = speech_type_data[speech_type_data['ParticipantID'].isin(opt_participants)].reset_index(drop=True)
    listener_holdout = speech_type_data[speech_type_data['ParticipantID'].isin(holdout_participants)].reset_index(drop=True)
    print(f"Optimization set count: {listener_opt.shape[0]}")
    print(f"Held-out set count: {listener_holdout.shape[0]}")
    
    # Verify that both splits contain all vowels
    print("\nVerifying vowel distribution in optimization and held-out sets:")
    print("Listener Optimization Vowel Counts:")
    print(listener_opt['Response.Vowel'].value_counts())
    print("\nListener Held-out Vowel Counts:")
    print(listener_holdout['Response.Vowel'].value_counts())
    
    # ------------------ Cross-Validation Setup ------------------
    
    # Cross-validation setup for phonetic_data (5-fold split)
    print("\nSetting up cross-validation for phonetic data...")
    group_kfold = GroupKFold(n_splits=5)
    if 'Talker' not in phonetic_df.columns:
        raise KeyError("Column 'Talker' not found in phonetic_data.")
    groups = phonetic_df['Talker']
    print("Cross-validation setup complete.")
    
    # ------------------ Model Parameters ------------------
    
    # Define unique categories as the intersection of phonetic and perceptual vowels
    unique_categories = sorted(set(phonetic_df['Vowel'].unique()).intersection(set(perceptual_df['Response.Vowel'].unique())))
    prior = {cat: 1.0 for cat in unique_categories}  # Uniform prior
    print(f"\nUnique categories: {unique_categories}")
    
    # ------------------ Optimization ------------------
    
    # Combined objective function for optimization
    def objective_function(params):
        """
        Objective function to be minimized (negative mean tuning log likelihood).
        
        Parameters:
        - params: Array containing [weight_f1, weight_f2, scaling]
        
        Returns:
        - Negative mean tuning log likelihood
        """
        weight_f1, weight_f2, scaling = params
        # Constraint: weight_f1 + weight_f2 = 1
        if not np.isclose(weight_f1 + weight_f2, 1.0, atol=1e-4):
            return np.inf
        # Bounds already enforced by optimizer, but double-check
        if not (0.0 <= weight_f1 <= 1.0) or not (0.0 <= weight_f2 <= 1.0):
            return np.inf
        if scaling <= 0:
            return np.inf
        weights_array = np.array([weight_f1, weight_f2])
        mean_tune_ll = evaluate_model(
            train_data=phonetic_df,
            tune_data=listener_opt,
            scaling=scaling,
            group_kfold=group_kfold,
            weights=weights_array,
            distance_exponent=distance_exponent,
            prior=prior
        )
        # If mean_tune_ll is nan, return a large penalty
        if np.isnan(mean_tune_ll):
            return 1e6
        return -mean_tune_ll  # Negative for minimization
    
    # Callback function to print weights and scaling during optimization
    def optimization_callback(xk):
        weight_f1, weight_f2, scaling = xk
        print(f"Optimization iteration - Weights: F1={weight_f1:.4f}, F2={weight_f2:.4f}; Similarity Scaling: {scaling:.4f}")
    
    # Optimization bounds and constraints
    bounds = [
        (0.0, 1.0),  # weight_f1
        (0.0, 1.0),  # weight_f2
        (0.1, 1e5) # similarity_scaling (set a reasonable upper bound)
    ]
    initial_guess = [0.5, 0.5, 10.0]  # Initial scaling set to 10.0
    noise_level = 0.0  # No noise in the formants for now
    constraints = {
        'type': 'eq',
        'fun': lambda x: x[0] + x[1] - 1.0
    }
    
    # Define distance exponent and fixed sigma
    distance_exponent = 2
    fixed_sigma = 1.0  # Not used in modified similarity
    
    # Perform optimization
    print("\nStarting parameter optimization using scipy.optimize.minimize...")
    result = minimize(
        objective_function,
        initial_guess,
        method='SLSQP',
        bounds=bounds,
        constraints=constraints,
        options={'disp': True, 'maxiter': 90},
        callback=optimization_callback
    )
    
    if result.success:
        optimized_weight_f1, optimized_weight_f2, optimized_scaling = result.x
        print(f"\nOptimization successful!")
        print(f"Optimized Feature Weight F1: {optimized_weight_f1:.4f}")
        print(f"Optimized Feature Weight F2: {optimized_weight_f2:.4f}")
        print(f"Optimized Similarity Scaling: {optimized_scaling:.4f}")
    else:
        print("\nOptimization failed.")
        print(result.message)
        # Assign default or initial values if optimization fails
        optimized_weight_f1, optimized_weight_f2, optimized_scaling = initial_guess
        print(f"Using initial weights and scaling: F1={optimized_weight_f1}, F2={optimized_weight_f2}, Scaling={optimized_scaling}")
    
    # ------------------ Overfitting Estimation ------------------
    
    print("\nEstimating overfitting by evaluating on held-out listener data...")
    
    # Train exemplars on the entire phonetic_df
    exemplars_test = defaultdict(list)
    for _, row in phonetic_df.iterrows():
        category = row['Vowel']
        features = row[['F1_z', 'F2_z']].tolist()
        noisy = add_noise(features, noise_level)
        if not np.any(np.isnan(noisy)) and not np.any(np.isinf(noisy)):
            exemplars_test[category].append(noisy)
    
    test_labels = []
    test_posteriors = []
    
    for _, row in listener_holdout.iterrows():
        true = row['Response.Vowel']
        input_feat = row[['F1_z', 'F2_z']].tolist()
        noisy_input = add_noise(input_feat, noise_level)
        if not np.any(np.isnan(noisy_input)) and not np.any(np.isinf(noisy_input)):
            posteriors = predict_posterior(
                noisy_input,
                exemplars_test,
                np.array([optimized_weight_f1, optimized_weight_f2]),
                distance_exponent,
                optimized_scaling,
                prior
            )
            if posteriors:
                test_posteriors.append(posteriors)
                test_labels.append(true)
    
    if test_labels:
        test_log_likelihood = sum([
            calculate_log_likelihood(t, p, unique_categories)
            for t, p in zip(test_labels, test_posteriors)
        ])
        print(f"Held-out Listener Log Likelihood: {test_log_likelihood:.4f}")
    else:
        print("No valid held-out listener predictions.")
    
    # ------------------ Vowel Space Boundary Visualization ------------------
    
    print("\nVisualizing vowel space boundaries...")
    
    # Define grid for visualization using z-scored values
    f2_range = np.linspace(phonetic_df['F2_z'].min() - 1, phonetic_df['F2_z'].max() + 1, 100)
    f1_range = np.linspace(phonetic_df['F1_z'].min() - 1, phonetic_df['F1_z'].max() + 1, 100)
    F1, F2 = np.meshgrid(f1_range, f2_range)
    grid_points = np.c_[F1.ravel(), F2.ravel()]
    
    # Predict posteriors for grid points
    grid_posteriors = []
    for point in grid_points:
        post = predict_posterior(
            point.tolist(),
            exemplars_test,
            np.array([optimized_weight_f1, optimized_weight_f2]),
            distance_exponent,
            optimized_scaling,
            prior
        )
        if post:
            grid_posteriors.append(post)
        else:
            grid_posteriors.append({cat: 0 for cat in unique_categories})
    
    # Assign most probable vowel
    predicted_vowels = [max(p, key=p.get) if p else "None" for p in grid_posteriors]
    
    # Map vowel categories to IPA symbols
    ipa_mapping = {
        'uw1': 'u',
        'ih1': 'ɪ',
        'uh1': 'ʊ',
        'iy1': 'i',
        'ae1': 'æ',
        'aa1': 'ɑ',
        'ah1': 'ʌ',
        'eh1': 'ɛ'
    }
    predicted_vowels_ipa = [ipa_mapping.get(vowel, 'None') for vowel in predicted_vowels]
    
    # Create integer mapping for color coding
    unique_ipa = sorted(set(ipa_mapping.values()))
    vowel_to_int = {vowel: idx for idx, vowel in enumerate(unique_ipa)}
    predicted_vowels_int = np.array([vowel_to_int.get(vowel, -1) for vowel in predicted_vowels_ipa]).reshape(F1.shape)
    
    # Plotting with numerical values
    plt.figure(figsize=(10, 8))
    contour = plt.contourf(F2, F1, predicted_vowels_int, alpha=0.5, levels=len(unique_ipa)+1, cmap='tab20')
    plt.title(f'Vowel Space Boundary for {model_name}')
    plt.xlabel('F2')
    plt.ylabel('F1')
    
    # Create custom legend for IPA symbols
    handles = []
    labels = []
    for idx, vowel in enumerate(unique_ipa):
        handles.append(plt.Line2D([], [], marker="s", linestyle="", color=plt.cm.tab20(idx / len(unique_ipa))))
        labels.append(f'[{vowel}]')
    plt.legend(handles, labels, title='Vowel Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    plt.colorbar(contour, ticks=range(len(unique_ipa)), label='Vowel Categories')
    plt.tight_layout()
    plt.show()
    
    # ------------------ Categorization Predictions Visualization ------------------
    
    print("\nVisualizing categorization predictions over F1-F2 space...")
    
    # Plot listener optimization points with boundaries and log-likelihood shading
    plt.figure(figsize=(10, 8))
    plt.contourf(F2, F1, predicted_vowels_int, alpha=0.3, levels=len(unique_ipa)+1, cmap='tab20')
    
    # Calculate mean log likelihood per listener point
    listener_opt['mean_log_ll'] = listener_opt.apply(
        lambda row: calculate_log_likelihood(
            row['Response.Vowel'],
            predict_posterior(
                row[['F1_z', 'F2_z']].tolist(),
                exemplars_test,
                np.array([optimized_weight_f1, optimized_weight_f2]),
                distance_exponent,
                optimized_scaling,
                prior
            ),
            unique_categories
        ),
        axis=1
    )
    
    # Plot listener optimization points with mean log likelihood
    scatter = plt.scatter(listener_opt['F2_z'], listener_opt['F1_z'], 
                          c=listener_opt['mean_log_ll'], cmap='gray', edgecolor='k', alpha=0.7)
    plt.colorbar(scatter, label='Mean Log Likelihood')
    plt.title(f'Categorization Predictions for {model_name}')
    plt.xlabel('F2')
    plt.ylabel('F1')
    
    # Create custom legend for IPA symbols
    plt.legend(handles, labels, title='Vowel Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    # ------------------ Store Model Results ------------------
    
    model_results[model_name] = {
        'optimized_weights': [optimized_weight_f1, optimized_weight_f2],
        'optimized_scaling': optimized_scaling,
        'held_out_log_likelihood': test_log_likelihood if test_labels else None
    }

# ------------------ Comparison Visualization Between Models ------------------

print("\nComparing all models...")

# Ensure all models are available
required_models = [
    'MODEL-1.1 - Natural',
    'MODEL-1.2 - Synthetic',
    'MODEL-2.1 - Natural',
    'MODEL-2.2 - Synthetic'
]

if all(model in model_results for model in required_models):
    # Compare Natural vs Synthetic
    for speech_type in ['Natural', 'Synthetic']:
        if speech_type == 'Natural':
            model1_name = 'MODEL-1.1 - Natural'
            model2_name = 'MODEL-2.1 - Natural'
        else:
            model1_name = 'MODEL-1.2 - Synthetic'
            model2_name = 'MODEL-2.2 - Synthetic'
        
        ll1 = model_results[model1_name]['held_out_log_likelihood']
        ll2 = model_results[model2_name]['held_out_log_likelihood']
        
        plt.figure(figsize=(6, 6))
        categories = ['Model-1', 'Model-2']
        log_likelihoods = [ll1, ll2]
        sns.barplot(x=categories, y=log_likelihoods, palette='viridis')
        plt.title(f'Held-out Listener Log Likelihood Comparison for {speech_type} Speech')
        plt.ylabel('Log Likelihood')
        plt.xlabel('Model')
        plt.tight_layout()
        plt.show()
else:
    print("Not all required models are available for comparison.")
```

```python
# FILE: Untitled5_v2.ipynb

import numpy as np
import pandas as pd
from scipy.spatial.distance import euclidean
from collections import defaultdict
from sklearn.model_selection import GroupKFold, StratifiedKFold
from sklearn.metrics import log_loss, accuracy_score
import random
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import LabelBinarizer
from itertools import product
from scipy.stats import zscore
import plotly.express as px
from scipy.optimize import minimize
import seaborn as sns

# ------------------ CCure Normalization Function ------------------

def ccure_normalize(df, feature_cols):
    """
    Perform CCure Normalization on specified feature columns by removing talker-specific means.
    
    Parameters:
    - df: pandas DataFrame to normalize.
    - feature_cols: List of column names to apply normalization.
    
    Returns:
    - Normalized DataFrame.
    """
    for col in feature_cols:
        df[col] = df.groupby('Talker')[col].transform(lambda x: x - x.mean())
    return df

# ------------------ Helper Functions ------------------

def add_noise(features, noise_level):
    """Add Gaussian noise to features based on noise_level."""
    return [f + np.random.normal(0, noise_level * abs(f)) for f in features]

def compute_distance(input_features, exemplar_features, weights, exponent):
    """Compute weighted distance between input and exemplar features."""
    weighted_diff = weights * (np.array(input_features) - np.array(exemplar_features))
    distance = np.sum(np.abs(weighted_diff) ** exponent) ** (1/exponent)
    return distance

def compute_similarity(distance, scaling):
    """Compute similarity based on the modified similarity function."""
    return np.exp(-distance ** scaling)

def predict_posterior(input_features, exemplars, weights, exponent, scaling, prior):
    """Predict posterior probabilities for each category."""
    category_similarities = {}
    total_similarity = 0
    for category, ex_list in exemplars.items():
        sim_sum = sum(
            compute_similarity(
                compute_distance(input_features, ex, weights, exponent),
                scaling
            )
            for ex in ex_list
        )
        category_similarities[category] = sim_sum
        total_similarity += sim_sum
    if total_similarity == 0:
        return None
    posteriors = {cat: (sim_sum * prior.get(cat, 1.0)) for cat, sim_sum in category_similarities.items()}
    total_posterior = sum(posteriors.values())
    if total_posterior == 0:
        return None
    posteriors = {cat: (sim / total_posterior) for cat, sim in posteriors.items()}
    return posteriors

def calculate_log_likelihood(true_label, posterior, classes):
    """Calculate log likelihood for a single prediction."""
    if posterior is None:
        return 0
    prob = posterior.get(true_label, 1e-10)
    prob = max(prob, 1e-10)
    return np.log(prob)

def evaluate_model(train_data, tune_data, scaling, group_kfold, weights, distance_exponent, prior):
    """
    Evaluate the model by computing the mean tuning log likelihood across cross-validation folds.
    
    Parameters:
    - train_data: Training dataset (phonetic data)
    - tune_data: Tuning dataset (perceptual optimization fold)
    - scaling: Scaling factor for similarity computation
    - group_kfold: GroupKFold splitter for cross-validation
    - weights: Array of feature weights [weight_f1, weight_f2]
    - distance_exponent: Exponent for distance computation
    - prior: Prior probabilities for each category
    
    Returns:
    - mean_tune_ll: Mean tuning log likelihood across all cross-validation folds
    """
    print(f"Evaluating model with weights=F1: {weights[0]:.8f}, F2: {weights[1]:.8f}; similarity_scaling: {scaling:.8f}")
    tune_log_likelihoods = []
    fold = 1
    for train_idx, _ in group_kfold.split(train_data, groups=train_data['Talker']):
        print(f"  Processing talker fold {fold}...")
        cv_train = train_data.iloc[train_idx].reset_index(drop=True)
        
        # Train exemplars on cv_train
        exemplars_train = defaultdict(list)
        for _, row in cv_train.iterrows():
            category = row['Vowel']
            features = row[['F1_z', 'F2_z']].tolist()
            noisy = add_noise(features, noise_level)
            if not np.any(np.isnan(noisy)) and not np.any(np.isinf(noisy)):
                exemplars_train[category].append(noisy)
        
        # Evaluate on tuning data
        tune_labels = []
        tune_posteriors = []
        
        for _, row in tune_data.iterrows():
            true = row['Response.Vowel']
            input_feat = row[['F1_z', 'F2_z']].tolist()
            noisy_input = add_noise(input_feat, noise_level)
            if not np.any(np.isnan(noisy_input)) and not np.any(np.isinf(noisy_input)):
                posteriors = predict_posterior(
                    noisy_input, exemplars_train, weights, distance_exponent, scaling, prior
                )
                if posteriors:
                    tune_posteriors.append(posteriors)
                    tune_labels.append(true)
        
        if tune_labels:
            fold_tune_log_likelihood = sum(
                [calculate_log_likelihood(t, p, unique_categories) for t, p in zip(tune_labels, tune_posteriors)]
            )
            tune_log_likelihoods.append(fold_tune_log_likelihood)
            print(f"  Fold {fold} - Tuning Log Likelihood: {fold_tune_log_likelihood:.4f}")
        else:
            print(f"  Fold {fold} - No valid tuning predictions.")
        fold += 1
    
    if tune_log_likelihoods:
        mean_tune_ll = np.mean(tune_log_likelihoods)
        print(f"  Mean Tuning Log Likelihood: {mean_tune_ll:.4f}")
    else:
        mean_tune_ll = np.nan
        print("  No valid tuning log likelihoods across folds.")
    
    return mean_tune_ll

# ------------------ Visualization Functions ------------------

def plot_vowel_space_boundary(model_name, unique_ipa, ipa_mapping, vowel_to_int, F1, F2, predicted_vowels_int):
    """
    Plots the vowel space boundary for a given model with consistent color mapping.

    Parameters:
    - model_name (str): Name of the model.
    - unique_ipa (list): List of unique IPA symbols.
    - ipa_mapping (dict): Mapping from vowel categories to IPA symbols.
    - vowel_to_int (dict): Mapping from IPA symbols to integer indices.
    - F1 (np.ndarray): 2D grid array for F1 values.
    - F2 (np.ndarray): 2D grid array for F2 values.
    - predicted_vowels_int (np.ndarray): 2D array of integer-mapped predicted vowels.
    """
    # Define consistent colors using ListedColormap
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_ipa)))
    cmap = ListedColormap(colors)

    plt.figure(figsize=(15, 12))  # Increased figure size

    # Plot the vowel space boundary
    contour = plt.contourf(F2, F1, predicted_vowels_int, alpha=0.5, levels=len(unique_ipa)+1, cmap=cmap)
    plt.title(f'Vowel Space Boundary for {model_name}', fontsize=20)
    plt.xlabel('F2', fontsize=16)
    plt.ylabel('F1', fontsize=16)

    # Create custom legend with matching colors
    handles = []
    labels = []
    for idx, vowel in enumerate(unique_ipa):
        handles.append(plt.Line2D([], [], marker="s", linestyle="", color=colors[idx]))
        labels.append(f'[{vowel}]')
    plt.legend(handles, labels, title='Vowel Categories', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14, title_fontsize=16)

    # Add colorbar
    cbar = plt.colorbar(contour, ticks=range(len(unique_ipa)), label='Vowel Categories')
    cbar.ax.set_yticklabels([f'[{vowel}]' for vowel in unique_ipa])

    plt.tight_layout()
    plt.show()

def plot_categorization_predictions(model_name, unique_ipa, ipa_mapping, vowel_to_int, F1, F2, predicted_vowels_int, listener_opt):
    """
    Plots categorization predictions over F1-F2 space with listener optimization points.

    Parameters:
    - model_name (str): Name of the model.
    - unique_ipa (list): List of unique IPA symbols.
    - ipa_mapping (dict): Mapping from vowel categories to IPA symbols.
    - vowel_to_int (dict): Mapping from IPA symbols to integer indices.
    - F1 (np.ndarray): 2D grid array for F1 values.
    - F2 (np.ndarray): 2D grid array for F2 values.
    - predicted_vowels_int (np.ndarray): 2D array of integer-mapped predicted vowels.
    - listener_opt (pd.DataFrame): DataFrame containing listener optimization data with 'F1_z', 'F2_z', and 'mean_log_ll'.
    """
    # Define consistent colors using ListedColormap
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_ipa)))
    cmap = ListedColormap(colors)

    plt.figure(figsize=(15, 12))  # Increased figure size

    # Plot the vowel space boundary
    plt.contourf(F2, F1, predicted_vowels_int, alpha=0.3, levels=len(unique_ipa)+1, cmap=cmap)

    # Create custom legend with matching colors
    handles = []
    labels = []
    for idx, vowel in enumerate(unique_ipa):
        handles.append(plt.Line2D([], [], marker="s", linestyle="", color=colors[idx]))
        labels.append(f'[{vowel}]')
    plt.legend(handles, labels, title='Vowel Categories', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14, title_fontsize=16)

    # Plot listener optimization points with mean log likelihood
    scatter = plt.scatter(
        listener_opt['F2_z'], 
        listener_opt['F1_z'], 
        c=listener_opt['mean_log_ll'], 
        cmap='gray', 
        edgecolor='k', 
        alpha=0.7
    )
    plt.colorbar(scatter, label='Mean Log Likelihood')

    plt.title(f'Categorization Predictions for {model_name}', fontsize=20)
    plt.xlabel('F2', fontsize=16)
    plt.ylabel('F1', fontsize=16)

    plt.tight_layout()
    plt.show()

# ------------------ Data Loading and Preprocessing ------------------

print("Loading datasets...")
# Set low_memory=False to suppress DtypeWarning
phonetic_data = pd.read_csv(
    '/scratch/ashar58/RA_Florian/Code/Li_Xie_2020_L1_vowels_statistics_general.csv',
    low_memory=False
)
perceptual_data = pd.read_csv(
    "/scratch/ashar58/RA_Florian/Code/Experiment-NORM-AB-after-exclusions.csv",
    low_memory=False
)
print("Datasets loaded successfully.")

# Modify 'Response.Vowel' in perceptual_data
vowel_mapping = {
    '[u]': 'uw1',
    '[ɪ]': 'ih1',
    '[ʊ]': 'uh1',
    '[i]': 'iy1',
    '[æ]': 'ae1',
    '[ɑ]': 'aa1',
    '[ʌ]': 'ah1',  
    '[ɛ]': 'eh1',
}
perceptual_data['Response.Vowel'] = perceptual_data['Response.Vowel'].map(vowel_mapping)
print("Perceptual vowels mapped successfully.")

# Remove rows with NA in 'Response.Vowel'
initial_perc_rows = perceptual_data.shape[0]
perceptual_data = perceptual_data.dropna(subset=['Response.Vowel'])
final_perc_rows = perceptual_data.shape[0]
rows_dropped = initial_perc_rows - final_perc_rows
print(f"Removed {rows_dropped} rows with NA in 'Response.Vowel'.")

# Define feature columns
phonetic_feature_cols = ['F1_gm', 'F2_gm']  # Phonetic features in Hz
perceptual_feature_cols = ['F1_Hz_r', 'F2_Hz_r']  # Perceptual features in Hz

# Compute z-scores for visualization
print("Computing z-scores for visualization...")
# Combine F1 from phonetic and perceptual data
all_F1 = pd.concat([phonetic_data['F1_gm'], perceptual_data['F1_Hz_r']], axis=0)
all_F2 = pd.concat([phonetic_data['F2_gm'], perceptual_data['F2_Hz_r']], axis=0)
combined = pd.DataFrame({
    'F1_z': zscore(all_F1),
    'F2_z': zscore(all_F2)
})

# Assign z-scores using positional indexing
phonetic_data[['F1_z', 'F2_z']] = combined.iloc[:len(phonetic_data)][['F1_z', 'F2_z']].values
perceptual_data[['F1_z', 'F2_z']] = combined.iloc[len(phonetic_data):][['F1_z', 'F2_z']].values
print("Z-scores computed for visualization.")

# ------------------ Model Definitions ------------------

models = {
    'MODEL-1.1 - Natural': {
        'normalize': False,
        'speech_type': 'natural'
    },
    'MODEL-1.2 - Synthetic': {
        'normalize': False,
        'speech_type': 'synthetic'
    },
    'MODEL-2.1 - Natural': {
        'normalize': True,
        'speech_type': 'natural'
    },
    'MODEL-2.2 - Synthetic': {
        'normalize': True,
        'speech_type': 'synthetic'
    }
}

# Initialize dictionaries to store model results
model_results = {}

# ------------------ Iterate Over Models ------------------

for model_name, model_params in models.items():
    print(f"\n----- Running {model_name} -----")
    
    # Copy datasets to avoid modifying originals
    phonetic_df = phonetic_data.copy()
    perceptual_df = perceptual_data.copy()
    
    # Apply CCure Normalization if required
    if model_params['normalize']:
        print("Applying CCure Normalization...")
        phonetic_df = ccure_normalize(phonetic_df, phonetic_feature_cols)
        perceptual_df = ccure_normalize(perceptual_df, perceptual_feature_cols)
        print("CCure Normalization applied.")
    else:
        print("Skipping CCure Normalization.")
    
    # Select the appropriate speech type data
    print("Splitting perceptual data into natural and synthetic based on 'Experiment' column...")
    if 'Experiment' not in perceptual_df.columns:
        raise KeyError("Column 'Experiment' not found in perceptual_data.")
    natural_data = perceptual_df[perceptual_df['Experiment'] == 'Experiment 1a (natural)'].reset_index(drop=True)
    synthetic_data = perceptual_df[perceptual_df['Experiment'] == 'Experiment 1b (synthesized)'].reset_index(drop=True)
    print(f"Natural data count: {natural_data.shape[0]}")
    print(f"Synthetic data count: {synthetic_data.shape[0]}")
    
    if model_params['speech_type'] == 'natural':
        speech_type_data = natural_data
    elif model_params['speech_type'] == 'synthetic':
        speech_type_data = synthetic_data
    else:
        raise ValueError("Invalid speech type specified.")
    
    # Split speech_type_data into optimization and held-out sets based on 'ParticipantID'
    print("Splitting speech type data into optimization and held-out sets based on 'ParticipantID'...")
    if 'ParticipantID' not in speech_type_data.columns:
        raise KeyError("Column 'ParticipantID' not found in speech_type_data.")
    unique_participants = speech_type_data['ParticipantID'].unique().tolist()
    random.shuffle(unique_participants)
    half = len(unique_participants) // 2
    opt_participants = unique_participants[:half]
    holdout_participants = unique_participants[half:]
    listener_opt = speech_type_data[speech_type_data['ParticipantID'].isin(opt_participants)].reset_index(drop=True)
    listener_holdout = speech_type_data[speech_type_data['ParticipantID'].isin(holdout_participants)].reset_index(drop=True)
    print(f"Optimization set count: {listener_opt.shape[0]}")
    print(f"Held-out set count: {listener_holdout.shape[0]}")
    
    # Verify that both splits contain all vowels
    print("\nVerifying vowel distribution in optimization and held-out sets:")
    print("Listener Optimization Vowel Counts:")
    print(listener_opt['Response.Vowel'].value_counts())
    print("\nListener Held-out Vowel Counts:")
    print(listener_holdout['Response.Vowel'].value_counts())
    
    # ------------------ Cross-Validation Setup ------------------
    
    # Cross-validation setup for phonetic_data (5-fold split)
    print("\nSetting up cross-validation for phonetic data...")
    group_kfold = GroupKFold(n_splits=5)
    if 'Talker' not in phonetic_df.columns:
        raise KeyError("Column 'Talker' not found in phonetic_data.")
    groups = phonetic_df['Talker']
    print("Cross-validation setup complete.")
    
    # ------------------ Model Parameters ------------------
    
    # Define unique categories as the intersection of phonetic and perceptual vowels
    unique_categories = sorted(set(phonetic_df['Vowel'].unique()).intersection(set(perceptual_df['Response.Vowel'].unique())))
    prior = {cat: 1.0 for cat in unique_categories}  # Uniform prior
    print(f"\nUnique categories: {unique_categories}")
    
    # ------------------ Optimization ------------------
    
    # Combined objective function for optimization
    def objective_function(params):
        """
        Objective function to be minimized (negative mean tuning log likelihood).
        
        Parameters:
        - params: Array containing [weight_f1, weight_f2, scaling]
        
        Returns:
        - Negative mean tuning log likelihood
        """
        weight_f1, weight_f2, scaling = params
        # Constraint: weight_f1 + weight_f2 = 1
        if not np.isclose(weight_f1 + weight_f2, 1.0, atol=1e-4):
            return np.inf
        # Bounds already enforced by optimizer, but double-check
        if not (0.0 <= weight_f1 <= 1.0) or not (0.0 <= weight_f2 <= 1.0):
            return np.inf
        if scaling <= 0:
            return np.inf
        weights_array = np.array([weight_f1, weight_f2])
        mean_tune_ll = evaluate_model(
            train_data=phonetic_df,
            tune_data=listener_opt,
            scaling=scaling,
            group_kfold=group_kfold,
            weights=weights_array,
            distance_exponent=distance_exponent,
            prior=prior
        )
        # If mean_tune_ll is nan, return a large penalty
        if np.isnan(mean_tune_ll):
            return 1e6
        return -mean_tune_ll  # Negative for minimization
    
    # Callback function to print weights and scaling during optimization
    def optimization_callback(xk):
        weight_f1, weight_f2, scaling = xk
        print(f"Optimization iteration - Weights: F1={weight_f1:.4f}, F2={weight_f2:.4f}; Similarity Scaling: {scaling:.4f}")
    
    # Optimization bounds and constraints
    bounds = [
        (0.0, 1.0),  # weight_f1
        (0.0, 1.0),  # weight_f2
        (0.1, 1e5) # similarity_scaling (set a reasonable upper bound)
    ]
    initial_guess = [0.5, 0.5, 10.0]  # Initial scaling set to 10.0
    noise_level = 0.0  # No noise in the formants for now
    constraints = {
        'type': 'eq',
        'fun': lambda x: x[0] + x[1] - 1.0
    }
    
    # Define distance exponent and fixed sigma
    distance_exponent = 2
    fixed_sigma = 1.0  # Not used in modified similarity
    
    # Perform optimization
    print("\nStarting parameter optimization using scipy.optimize.minimize...")
    result = minimize(
        objective_function,
        initial_guess,
        method='SLSQP',
        bounds=bounds,
        constraints=constraints,
        options={'disp': True, 'maxiter': 90},
        callback=optimization_callback
    )
    
    if result.success:
        optimized_weight_f1, optimized_weight_f2, optimized_scaling = result.x
        print(f"\nOptimization successful!")
        print(f"Optimized Feature Weight F1: {optimized_weight_f1:.4f}")
        print(f"Optimized Feature Weight F2: {optimized_weight_f2:.4f}")
        print(f"Optimized Similarity Scaling: {optimized_scaling:.4f}")
    else:
        print("\nOptimization failed.")
        print(result.message)
        # Assign default or initial values if optimization fails
        optimized_weight_f1, optimized_weight_f2, optimized_scaling = initial_guess
        print(f"Using initial weights and scaling: F1={optimized_weight_f1}, F2={optimized_weight_f2}, Scaling={optimized_scaling}")
    
    # ------------------ Overfitting Estimation ------------------
    
    print("\nEstimating overfitting by evaluating on held-out listener data...")
    
    # Train exemplars on the entire phonetic_df
    exemplars_test = defaultdict(list)
    for _, row in phonetic_df.iterrows():
        category = row['Vowel']
        features = row[['F1_z', 'F2_z']].tolist()
        noisy = add_noise(features, noise_level)
        if not np.any(np.isnan(noisy)) and not np.any(np.isinf(noisy)):
            exemplars_test[category].append(noisy)
    
    test_labels = []
    test_posteriors = []
    
    for _, row in listener_holdout.iterrows():
        true = row['Response.Vowel']
        input_feat = row[['F1_z', 'F2_z']].tolist()
        noisy_input = add_noise(input_feat, noise_level)
        if not np.any(np.isnan(noisy_input)) and not np.any(np.isinf(noisy_input)):
            posteriors = predict_posterior(
                noisy_input,
                exemplars_test,
                np.array([optimized_weight_f1, optimized_weight_f2]),
                distance_exponent,
                optimized_scaling,
                prior
            )
            if posteriors:
                test_posteriors.append(posteriors)
                test_labels.append(true)
    
    if test_labels:
        test_log_likelihood = sum([
            calculate_log_likelihood(t, p, unique_categories)
            for t, p in zip(test_labels, test_posteriors)
        ])
        print(f"Held-out Listener Log Likelihood: {test_log_likelihood:.4f}")
    else:
        print("No valid held-out listener predictions.")
    
    # ------------------ Vowel Space Boundary Visualization ------------------
    
    print("\nVisualizing vowel space boundaries...")
    
    # Define grid for visualization using z-scored values
    f2_range = np.linspace(phonetic_df['F2_z'].min() - 1, phonetic_df['F2_z'].max() + 1, 100)
    f1_range = np.linspace(phonetic_df['F1_z'].min() - 1, phonetic_df['F1_z'].max() + 1, 100)
    F1_grid, F2_grid = np.meshgrid(f1_range, f2_range)
    grid_points = np.c_[F1_grid.ravel(), F2_grid.ravel()]
    
    # Predict posteriors for grid points
    grid_posteriors = []
    for point in grid_points:
        post = predict_posterior(
            point.tolist(),
            exemplars_test,
            np.array([optimized_weight_f1, optimized_weight_f2]),
            distance_exponent,
            optimized_scaling,
            prior
        )
        if post:
            grid_posteriors.append(post)
        else:
            grid_posteriors.append({cat: 0 for cat in unique_categories})
    
    # Assign most probable vowel
    predicted_vowels = [max(p, key=p.get) if p else "None" for p in grid_posteriors]
    
    # Map vowel categories to IPA symbols
    ipa_mapping = {
        'uw1': 'u',
        'ih1': 'ɪ',
        'uh1': 'ʊ',
        'iy1': 'i',
        'ae1': 'æ',
        'aa1': 'ɑ',
        'ah1': 'ʌ',
        'eh1': 'ɛ'
    }
    predicted_vowels_ipa = [ipa_mapping.get(vowel, 'None') for vowel in predicted_vowels]
    
    # Create integer mapping for color coding
    unique_ipa = sorted(set(ipa_mapping.values()))
    vowel_to_int = {vowel: idx for idx, vowel in enumerate(unique_ipa)}
    predicted_vowels_int = np.array([vowel_to_int.get(vowel, -1) for vowel in predicted_vowels_ipa]).reshape(F1_grid.shape)
    
    # Plotting Vowel Space Boundary
    plot_vowel_space_boundary(
        model_name=model_name,
        unique_ipa=unique_ipa,
        ipa_mapping=ipa_mapping,
        vowel_to_int=vowel_to_int,
        F1=F1_grid,
        F2=F2_grid,
        predicted_vowels_int=predicted_vowels_int
    )
    
    # ------------------ Categorization Predictions Visualization ------------------
    
    print("\nVisualizing categorization predictions over F1-F2 space...")
    
    # Calculate mean log likelihood per listener point
    listener_opt['mean_log_ll'] = listener_opt.apply(
        lambda row: calculate_log_likelihood(
            row['Response.Vowel'],
            predict_posterior(
                row[['F1_z', 'F2_z']].tolist(),
                exemplars_test,
                np.array([optimized_weight_f1, optimized_weight_f2]),
                distance_exponent,
                optimized_scaling,
                prior
            ),
            unique_categories
        ),
        axis=1
    )
    
    # Plot Categorization Predictions
    plot_categorization_predictions(
        model_name=model_name,
        unique_ipa=unique_ipa,
        ipa_mapping=ipa_mapping,
        vowel_to_int=vowel_to_int,
        F1=F1_grid,
        F2=F2_grid,
        predicted_vowels_int=predicted_vowels_int,
        listener_opt=listener_opt
    )
    
    # ------------------ Store Model Results ------------------
    
    model_results[model_name] = {
        'optimized_weights': [optimized_weight_f1, optimized_weight_f2],
        'optimized_scaling': optimized_scaling,
        'held_out_log_likelihood': test_log_likelihood if test_labels else None,
        'predicted_vowels_int': predicted_vowels_int,
        'listener_opt': listener_opt.copy()
    }

# ------------------ Comparison Visualization Between Models ------------------

print("\nComparing all models...")

# Ensure all models are available
required_models = [
    'MODEL-1.1 - Natural',
    'MODEL-1.2 - Synthetic',
    'MODEL-2.1 - Natural',
    'MODEL-2.2 - Synthetic'
]

if all(model in model_results for model in required_models):
    # Compare Natural vs Synthetic
    for speech_type in ['Natural', 'Synthetic']:
        if speech_type == 'Natural':
            model1_name = 'MODEL-1.1 - Natural'
            model2_name = 'MODEL-2.1 - Natural'
        else:
            model1_name = 'MODEL-1.2 - Synthetic'
            model2_name = 'MODEL-2.2 - Synthetic'
        
        ll1 = model_results[model1_name]['held_out_log_likelihood']
        ll2 = model_results[model2_name]['held_out_log_likelihood']
        
        plt.figure(figsize=(10, 8))  # Increased figure size
        categories = ['Model-1', 'Model-2']
        log_likelihoods = [ll1, ll2]
        sns.barplot(x=categories, y=log_likelihoods, palette='viridis')
        plt.title(f'Held-out Listener Log Likelihood Comparison for {speech_type} Speech', fontsize=16)
        plt.ylabel('Log Likelihood', fontsize=14)
        plt.xlabel('Model', fontsize=14)
        plt.tight_layout()
        plt.show()
else:
    print("Not all required models are available for comparison.")
```

```python
# FILE: Visualize_Vowel_Space.ipynb

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

# ------------------ Visualization Functions ------------------

def plot_vowel_space_boundary(model_name, unique_ipa, ipa_mapping, vowel_to_int, F1, F2, predicted_vowels_int):
    """
    Plots the vowel space boundary for a given model with consistent color mapping.

    Parameters:
    - model_name (str): Name of the model.
    - unique_ipa (list): List of unique IPA symbols.
    - ipa_mapping (dict): Mapping from vowel categories to IPA symbols.
    - vowel_to_int (dict): Mapping from IPA symbols to integer indices.
    - F1 (np.ndarray): 2D grid array for F1 values.
    - F2 (np.ndarray): 2D grid array for F2 values.
    - predicted_vowels_int (np.ndarray): 2D array of integer-mapped predicted vowels.
    """
    # Define consistent colors using ListedColormap
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_ipa)))
    cmap = ListedColormap(colors)

    plt.figure(figsize=(15, 12))  # Increased figure size

    # Plot the vowel space boundary
    contour = plt.contourf(F2, F1, predicted_vowels_int, alpha=0.5, levels=len(unique_ipa)+1, cmap=cmap)
    plt.title(f'Vowel Space Boundary for {model_name}', fontsize=20)
    plt.xlabel('F2', fontsize=16)
    plt.ylabel('F1', fontsize=16)

    # Create custom legend with matching colors
    handles = []
    labels = []
    for idx, vowel in enumerate(unique_ipa):
        handles.append(plt.Line2D([], [], marker="s", linestyle="", color=colors[idx]))
        labels.append(f'[{vowel}]')
    plt.legend(handles, labels, title='Vowel Categories', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14, title_fontsize=16)

    # Add colorbar
    cbar = plt.colorbar(contour, ticks=range(len(unique_ipa)), label='Vowel Categories')
    cbar.ax.set_yticklabels([f'[{vowel}]' for vowel in unique_ipa])

    plt.tight_layout()
    plt.show()

def plot_categorization_predictions(model_name, unique_ipa, ipa_mapping, vowel_to_int, F1, F2, predicted_vowels_int, listener_opt):
    """
    Plots categorization predictions over F1-F2 space with listener optimization points.

    Parameters:
    - model_name (str): Name of the model.
    - unique_ipa (list): List of unique IPA symbols.
    - ipa_mapping (dict): Mapping from vowel categories to IPA symbols.
    - vowel_to_int (dict): Mapping from IPA symbols to integer indices.
    - F1 (np.ndarray): 2D grid array for F1 values.
    - F2 (np.ndarray): 2D grid array for F2 values.
    - predicted_vowels_int (np.ndarray): 2D array of integer-mapped predicted vowels.
    - listener_opt (pd.DataFrame): DataFrame containing listener optimization data with 'F1_z', 'F2_z', and 'mean_log_ll'.
    """
    # Define consistent colors using ListedColormap
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_ipa)))
    cmap = ListedColormap(colors)

    plt.figure(figsize=(15, 12))  # Increased figure size

    # Plot the vowel space boundary
    plt.contourf(F2, F1, predicted_vowels_int, alpha=0.3, levels=len(unique_ipa)+1, cmap=cmap)

    # Create custom legend with matching colors
    handles = []
    labels = []
    for idx, vowel in enumerate(unique_ipa):
        handles.append(plt.Line2D([], [], marker="s", linestyle="", color=colors[idx]))
        labels.append(f'[{vowel}]')
    plt.legend(handles, labels, title='Vowel Categories', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=14, title_fontsize=16)

    # Plot listener optimization points with mean log likelihood
    scatter = plt.scatter(
        listener_opt['F2_z'], 
        listener_opt['F1_z'], 
        c=listener_opt['mean_log_ll'], 
        cmap='gray', 
        edgecolor='k', 
        alpha=0.7
    )
    plt.colorbar(scatter, label='Mean Log Likelihood')

    plt.title(f'Categorization Predictions for {model_name}', fontsize=20)
    plt.xlabel('F2', fontsize=16)
    plt.ylabel('F1', fontsize=16)

    plt.tight_layout()
    plt.show()

# ------------------ Load Stored Model Results ------------------

# # Assuming you have stored your model results in a pickle or JSON file.
# # Replace 'model_results.pkl' with your actual file path and format.
# import pickle

# # Load the stored model_results
# with open('model_results.pkl', 'rb') as file:
#     model_results = pickle.load(file)

# ------------------ Define Mappings ------------------

# Define vowel to IPA mapping
ipa_mapping = {
    'uw1': 'u',
    'ih1': 'ɪ',
    'uh1': 'ʊ',
    'iy1': 'i',
    'ae1': 'æ',
    'aa1': 'ɑ',
    'ah1': 'ʌ',
    'eh1': 'ɛ'
}

unique_ipa = sorted(set(ipa_mapping.values()))
vowel_to_int = {vowel: idx for idx, vowel in enumerate(unique_ipa)}

# ------------------ Generate and Plot Vowel Space Boundaries ------------------

for model_name, results in model_results.items():
    print(f"\n----- Visualizing {model_name} -----")
    
    # Extract stored results
    predicted_vowels_int = results.get('predicted_vowels_int')
    listener_opt = results.get('listener_opt')
    F1 = results.get('F1_grid')
    F2 = results.get('F2_grid')
    
    if predicted_vowels_int is None or listener_opt is None:
        print(f"Missing data for {model_name}. Skipping visualization.")
        continue
    
    # Plot Vowel Space Boundary
    plot_vowel_space_boundary(
        model_name=model_name,
        unique_ipa=unique_ipa,
        ipa_mapping=ipa_mapping,
        vowel_to_int=vowel_to_int,
        F1=F1,
        F2=F2,
        predicted_vowels_int=predicted_vowels_int
    )
    
    # Plot Categorization Predictions
    plot_categorization_predictions(
        model_name=model_name,
        unique_ipa=unique_ipa,
        ipa_mapping=ipa_mapping,
        vowel_to_int=vowel_to_int,
        F1=F1,
        F2=F2,
        predicted_vowels_int=predicted_vowels_int,
        listener_opt=listener_opt
    )

# ------------------ Comparison Visualization Between Models ------------------

print("\nComparing all models...")

# Ensure all required models are present
required_models = [
    'MODEL-1.1 - Natural',
    'MODEL-1.2 - Synthetic',
    'MODEL-2.1 - Natural',
    'MODEL-2.2 - Synthetic'
]

missing_models = [model for model in required_models if model not in model_results]
if missing_models:
    print(f"Missing models for comparison: {missing_models}")
else:
    for speech_type in ['Natural', 'Synthetic']:
        if speech_type == 'Natural':
            model1 = 'MODEL-1.1 - Natural'
            model2 = 'MODEL-2.1 - Natural'
        else:
            model1 = 'MODEL-1.2 - Synthetic'
            model2 = 'MODEL-2.2 - Synthetic'
        
        ll1 = model_results[model1].get('held_out_log_likelihood')
        ll2 = model_results[model2].get('held_out_log_likelihood')
        
        if ll1 is None or ll2 is None:
            print(f"Missing log likelihoods for {model1} or {model2}. Skipping comparison.")
            continue
        
        plt.figure(figsize=(10, 8))  # Increased figure size
        categories = ['Model-1', 'Model-2']
        log_likelihoods = [ll1, ll2]
        sns.barplot(x=categories, y=log_likelihoods, palette='viridis')
        plt.title(f'Held-out Listener Log Likelihood Comparison for {speech_type} Speech', fontsize=16)
        plt.ylabel('Log Likelihood', fontsize=14)
        plt.xlabel('Model', fontsize=14)
        plt.tight_layout()
        plt.show()
```
